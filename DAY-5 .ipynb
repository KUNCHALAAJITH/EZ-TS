# balanced paranthesis Expression

```python
stack=[]
s=input()
n=len(s)
check=0
flag=0
if n%2!=0:
    print("unbalanced expression")
else:
    for i in range(n):
        if s[i]=='[' or s[i]=='(' or s[i]=='{':
            stack.append(s[i])
            check+=1
        elif s[i]==']' or s[i]=='}' or s[i]==')':
            if (stack[-1]=='{' and s[i]=='}') or (stack[-1]=='[' and s[i]==']') or (stack[-1]=='(' and s[i]==')'):
                stack.pop()
                check-=1
            else:
                flag=1
                break
        else:
            flag=1
            break
    if flag==1 or check!=0:
        print("unbalanced Expression")
    else:
        print("balanced Expression")
```
